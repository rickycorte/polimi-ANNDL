```
from PIL import Image

# in this notebook we aim to train one dataset at time 

bipbip_dataset_dir = os.path.join(working_directory,'Development_Dataset/Training/Bipbip')

plants = ['Haricot','Mais']

bb_images = []
bb_masks = []

image_names = []
mask_names = []

for p in plants:
  image_names.append(os.listdir(os.path.join(bipbip_dataset_dir,p,'Images')))
  mask_names.append(os.listdir(os.path.join(bipbip_dataset_dir,p,'Masks')))

print(len(image_names))
print(len(mask_names))

print(image_names)
```

```
images_array_t = []
masks_array_t = []

target_size = 1024

c = 0
for p in plants:
  for i in image_names[c]:
    img = Image.open(os.path.join(bipbip_dataset_dir,p,'Images',i))
    img_array = np.array(img)
    img_array = np.resize(img_array,(target_size,target_size,3))
    images_array_t.append(img_array)
  c += 1

c = 0
for p in plants:
  for m in mask_names[c]:
    img = Image.open(os.path.join(bipbip_dataset_dir,p,'Masks',m))
    mask_arr = np.array(img)
    mask_arr = np.resize(img_array,(target_size,target_size,3))
    new_mask_arr = np.array(np.zeros((target_size,target_size,1),dtype = np.float32))
    new_mask_arr[np.where(np.all(mask_arr == [216, 124, 18], axis=-1))] = 0
    new_mask_arr[np.where(np.all(mask_arr == [255, 255, 255], axis=-1))] = 1
    new_mask_arr[np.where(np.all(mask_arr == [216, 67, 82], axis=-1))] = 2
    masks_array_t.append(new_mask_arr)
  c += 1

```
